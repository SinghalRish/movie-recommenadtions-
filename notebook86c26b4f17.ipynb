```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python


def convert3(text):
    L = []
    counter = 0
    for i in text:
        if counter !=3:
            L.append(i)
            counter += 1
        else:
            break  # Exit loop after processing 3 items
    return L
```

```python
movies['cast'][0]
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'][1]
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'][0]
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'].apply(stem)
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply(lambda x: x.lower())
new['tags'][0]
```

```python
!pip install scikit-learn

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') 
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out() 
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    #enumerate gives us list of tuples with disctances along with their indexes reverse = true means descending key = lambda x:x[1] means from list of
    #tuples sorting will be done on the baisi of index 1 ie distances 
    for i in distances[1:6]:  
        print(new.iloc[i[0]].title) 
        
    
```

```python
recommend('Avatar')
```
